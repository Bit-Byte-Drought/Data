```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel(
    '../Datasets/rainfall_District Wise Timeseries_2017.xls',                   #Relative path
    header = [1,2]                                                              #Selecting rows for column name
).replace('-', np.nan)                                                          #Replacing - with Nan
```

```python
df.head()
```

```python
df.apply(lambda x:x.isna().sum(), axis = 1).value_counts()
```

```python
df = df.iloc[:-1]
index = df[df.columns[0:2]]
index.columns = ['State', 'District']
df = df.drop(columns = df.columns[4::3])
df = df[df.columns[2:]]
df.index = pd.MultiIndex.from_frame(index)
df.head()
```

```python
Null = list()
df.apply(lambda x:Null.append(x.name) if x.isna().sum() == df.shape[1] else None,axis = 1)
Null = pd.DataFrame(Null)
Null.head()
```

```python
Null.to_csv('../Datasets/Null_Districts(2017).csv', index = False)
```

```python
import pandas as pd
import numpy as np
df = pd.read_excel(
    '../Datasets/rainfall_District Wise Timeseries_2013.xls',                   #Relative path
    header = [1,2]                                                              #Selecting rows for column name
).replace('-', np.nan)                                                          #Replacing - with Nan
```

```python
Actual = df[df.columns[2::3]].apply(lambda x:x.isna().sum())
```

As we can see above, all the attributes corresponding to `Actual` are only NaNs.

```python
Deviation = df[df.columns[4::3]].apply(lambda x:x.isna().sum())
```

```python
index = df[df.columns[0:2]]
index.columns = ['State', 'District']
df = df[df.columns[3::3]]
df.index = pd.MultiIndex.from_frame(index)
df.columns = pd.Series(df.columns.to_numpy()).apply(lambda x:x[0])
df.head()
```

Separating first two columns to a variable `index`.<br>
Renaming the columns from `index` in the second line.<br>
Considering only Normal attribute from all our days and saving it to `df`.<br>
Adding a MultiIndex to our df through `index`.<br>
Dropping the Recurring attribute name 'Normal(mm)'.

```python
Null = list()
df.apply(lambda x:Null.append(x.name) if x.isna().sum() == df.shape[1] else None,axis = 1)
Null = pd.DataFrame(Null)
Null.head()
```

```python
Null.to_csv('../Datasets/Null_Districts(2013).csv', index = False)
```
